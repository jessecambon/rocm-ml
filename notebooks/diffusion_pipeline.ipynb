## Inputs

```python
model_name = "stable-diffusion-v1-5/stable-diffusion-v1-5" # huggingface model to use with DiffusionPipeline
prompt = "A cartoon of a coffee snob making espresso with a lever machine" # image prompt
```

## Generate Image
Run cell below to generate the image

```python
from diffusers import DiffusionPipeline
import torch
from datetime import datetime

pipeline = DiffusionPipeline.from_pretrained(model_name, torch_dtype=torch.float16)
pipeline.to("cuda")
image = pipeline(prompt).images[0]
image
```

## Save Image
Run cell below to save the above image to disk

```python
dt_str = datetime.now().strftime("%Y-%m-%d-%H-%M-%S")
image.save(f"images/{dt_str}.png")
```
